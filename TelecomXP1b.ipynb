4.3 **Recuento de evasión por variables categóricas**

```python
def graficar_abandono(df, columnas, labels_dict=None, xmin=0, ymax=5000):
    """
    Grafica la distribución de abandono para múltiples columnas en subplots.

    Args:
        df: DataFrame con los datos completos
        columnas: Lista de columnas a graficar
        labels_dict: Diccionario con etiquetas para cada columna {columna: [label0, label1]}
        xmin: Valor mínimo del eje y
        ymax: Valor máximo del eje y
    """
    n_cols = len(columnas)
    fig, axes = plt.subplots(1, n_cols, figsize=(6*n_cols, 4))

    if n_cols == 1:
        axes = [axes]

    for idx, columna in enumerate(columnas):
        ax = axes[idx]
        abandono_serie = df.groupby(columna)['Abandono'].value_counts()
        abandono_df = abandono_serie.unstack(level='Abandono').fillna(0).astype(int)

        if labels_dict and columna in labels_dict:
            labels = [labels_dict[columna][0] if g == 0 else labels_dict[columna][1] for g in abandono_df.index]
        else:
            labels = ['No' if g == 0 else 'Sí' for g in abandono_df.index]

        x = np.arange(len(abandono_df))
        width = 0.35

        ax.bar(x - width/2, abandono_df.get(0, 0), width, label='No', color='C0')
        ax.bar(x + width/2, abandono_df.get(1, 0), width, label='Sí', color='C1')

        ax.set_title(f'Distribución de abandono por {columna}', fontsize=12)
        ax.set_xlabel(columna, fontsize=10)
        ax.set_xticks(x)
        ax.set_xticklabels(labels, fontsize=10)
        ax.set_ylim(bottom=xmin, top=ymax)
        ax.yaxis.set_ticks(ticks=[])
        ax.legend(title='Abandono', fontsize=10)


        offset = max(abandono_df.values.max() * 0.02, 5)
        for i in range(len(x)):
            val_no = abandono_df.get(0, 0).iloc[i]
            val_si = abandono_df.get(1, 0).iloc[i]
            total = val_no + val_si
            pct_no = val_no / total * 100 if total > 0 else 0
            pct_si = val_si / total * 100 if total > 0 else 0
            ax.text(x[i] - width/2, val_no + offset, f'{val_no} ({pct_no:.1f}%)', ha='center', fontsize=8)
            ax.text(x[i] + width/2, val_si + offset, f'{val_si} ({pct_si:.1f}%)', ha='center', fontsize=8)

        sns.despine(ax=ax, top=True, right=True, left=True)

    plt.tight_layout()
    sns.set_palette('Set2')
    plt.savefig(f'figuras/grafico_abandono_{"_".join(columnas)}.png')
    plt.show()
```

```python
def graficar_abandono_ohe(df, prefijos, titulos, labels_map=None, ymin=0, ymax=5000):
    """
    Grafica abandono para columnas One Hot Encoding con prefijos comunes en subplots.

    Args:
        df: DataFrame con los datos
        prefijos: Lista de prefijos de las columnas OHE (ej: ['Multilinea Telefónica', 'Internet'])
        titulos: Lista de títulos para cada subplot
        labels_map: Diccionario opcional para renombrar etiquetas del eje x
        ymin: Valor mínimo del eje y
        ymax: Valor máximo del eje y
    """
    n_plots = len(prefijos)
    n_cols = min(2, n_plots)
    n_rows = (n_plots + 1) // 2

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(8*n_cols, 4*n_rows))

    if n_plots == 1:
        axes = [axes]
    else:
        axes = axes.flatten()

    for idx, (prefijo, titulo) in enumerate(zip(prefijos, titulos)):
        ax = axes[idx]
        columnas = df.columns[df.columns.str.startswith(prefijo) | df.columns.isin(['Abandono'])]
        df_subset = df[columnas]
        ohe_cols = [col for col in columnas if col != 'Abandono']
        resultado = {}
        for col in ohe_cols:
            resultado[col] = df_subset[df_subset[col] == 1]['Abandono'].value_counts().sort_index()

        conteo_abandono = pd.DataFrame(resultado).T
        conteo_abandono.columns = ['No', 'Sí']

        if labels_map and prefijo in labels_map:
            conteo_abandono.index = [labels_map[prefijo].get(col, col) for col in conteo_abandono.index]
        else:
            conteo_abandono.index = [col.replace(f'{prefijo}: ', '') for col in conteo_abandono.index]

        x = np.arange(len(conteo_abandono))
        width = 0.35

        ax.bar(x - width/2, conteo_abandono['No'], width=width, label='No')
        ax.bar(x + width/2, conteo_abandono['Sí'], width=width, label='Sí')

        ax.set_title(f'Distribución de abandono por {titulo}', fontsize=14)
        ax.set_xlabel(titulo, fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels(conteo_abandono.index, fontsize=10)
        ax.set_ylim(bottom=ymin, top=ymax)
        ax.yaxis.set_ticks(ticks=[])
        ax.legend(title='Abandono', fontsize=10)

        offset = max(conteo_abandono.values.max() * 0.02, 5)
        for i in range(len(x)):
            val_no = conteo_abandono['No'].iloc[i]
            val_si = conteo_abandono['Sí'].iloc[i]
            total = val_no + val_si
            pct_no = val_no / total * 100 if total > 0 else 0
            pct_si = val_si / total * 100 if total > 0 else 0
            ax.text(x[i] - width/2, val_no + offset, f'{val_no} ({pct_no:.1f}%)', ha='center', fontsize=8)
            ax.text(x[i] + width/2, val_si + offset, f'{val_si} ({pct_si:.1f}%)', ha='center', fontsize=8)

        sns.despine(ax=ax, top=True, right=True, left=True)

    # Ocultar ejes vacíos si el número de plots es impar
    for idx in range(n_plots, len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    sns.set_palette('Set2')
    plt.savefig(f'figuras/grafico_abandono_{"_".join(titulos)}.png')
    plt.show()
```

6.0 **Insights Principales**

A partir del Análisis Exploratorio de Datos (EDA), se han identificado los siguientes factores como determinantes en la retención y abandono de clientes:

1. La antigüedad como factor crítico

Se evidencia una fuerte correlación negativa entre la antigüedad del cliente y la tasa de abandono. Los clientes con menor permanencia —particularmente durante los primeros seis meses— presentan una probabilidad significativamente mayor de cancelar el servicio.

A medida que aumenta la antigüedad, la tasa de fuga disminuye de manera sustancial, lo que sugiere un proceso progresivo de consolidación de la fidelidad. Este hallazgo posiciona la etapa inicial del ciclo de vida del cliente como un período estratégico para la implementación de acciones de retención temprana.

2. Impacto del método de pago

El análisis detecta una anomalía relevante asociada al método de pago “Cheque Electrónico”, el cual registra una tasa de abandono del 45,3%, cifra que duplica ampliamente la observada en otros medios de pago.

Este comportamiento sugiere la existencia de fricciones operativas, menor automatización o perfiles de clientes con menor vinculación financiera, constituyéndose en un factor de riesgo prioritario dentro de la estrategia de retención.

3. Vulnerabilidad en el segmento “Veteranos o Tercera edad”

El segmento demográfico identificado como “Veteranos” presenta una propensión al abandono superior al promedio general.

Si bien este grupo representa una proporción menor dentro de la base total de clientes, su elevada tasa de fuga exige un análisis segmentado más profundo, con el objetivo de identificar variables explicativas asociadas a necesidades específicas, sensibilidad al precio o brechas en la experiencia de servicio.

4. Tipo de contrato

Los clientes con modalidad contractual “Mes a Mes” exhiben una alta volatilidad en comparación con aquellos sujetos a contratos de uno o dos años.

La ausencia de compromiso de permanencia facilita la cancelación inmediata ante cualquier episodio de insatisfacción o frente a ofertas competitivas. En contraste, los contratos de largo plazo actúan como mecanismos de estabilización de la base de clientes, reduciendo la elasticidad de fuga.

5. Costo mensual y servicios adicionales

Se observa una relación positiva entre el nivel de cargo mensual y la probabilidad de abandono. Los clientes con planes de mayor valor —frecuentemente asociados a servicios de fibra óptica— muestran mayor exposición competitiva y, potencialmente, mayor sensibilidad al precio.

Adicionalmente, los clientes que no contratan servicios complementarios tales como Soporte Técnico, Seguridad en Línea o Backup presentan mayores tasas de fuga. Estos servicios adicionales operan como mecanismos de anclaje, incrementando el costo de cambio, la dependencia funcional y el valor percibido del servicio principal.

6. Facturación electrónica

Se identifica que los clientes adscritos a facturación electrónica presentan una tasa de abandono superior a aquellos que reciben facturación en formato físico.

Este comportamiento podría estar vinculado a un perfil de usuario más digital, con mayor acceso a información comparativa y mayor propensión a evaluar alternativas en línea. No obstante, este hallazgo requiere análisis complementario para descartar efectos de variables de confusión.

7.0 **Recomendaciones**

1. Programa “Primeros Pasos” (Retención Temprana)

a) Esquema de descuentos escalonados
En lugar de concentrar el incentivo comercial exclusivamente en el primer mes, se recomienda implementar un modelo de descuentos decrecientes durante los primeros seis meses de servicio.

Este diseño busca incentivar la permanencia progresiva del cliente hasta superar el denominado “umbral crítico de riesgo”, identificado en el análisis como el período de mayor probabilidad de abandono. La estructura escalonada permite alinear incentivos comerciales con objetivos de retención sostenida.

b) Onboarding con mejora temporal de prestaciones
Se propone ofrecer una mejora temporal en la velocidad contratada durante los primeros meses del servicio, como parte de una estrategia de generación acelerada de valor percibido.

Adicionalmente, se podría incorporar la opción de mantener dicha mejora de forma permanente mediante la suscripción a un contrato anual. Esta medida combina experiencia diferenciadora inicial con un mecanismo de conversión hacia mayor compromiso contractual.

2. Optimización y Migración de Métodos de Pago

a) Incentivo por automatización de pagos (autopago)
Se recomienda promover activamente la domiciliación de pagos mediante transferencias automáticas o tarjetas de crédito, ofreciendo beneficios tales como descuentos fijos o mejoras en el plan (por ejemplo, aumento de velocidad).

La automatización reduce la fricción operativa asociada al pago manual, disminuye la probabilidad de reconsideración mensual del servicio y contribuye a la estabilización del flujo de ingresos.

b) Rediseño estratégico de la facturación electrónica (“Alerta de Valor”)
Para los usuarios adscritos a facturación electrónica, se sugiere transformar el correo de notificación de pago en una herramienta de comunicación de valor.

En lugar de limitarse a un recordatorio transaccional, el mensaje podría incluir indicadores personalizados tales como consumo mensual de datos, comparativas de rendimiento, beneficios obtenidos y métricas de ahorro de tiempo o eficiencia.

Este enfoque refuerza el valor percibido del servicio y reduce la probabilidad de cancelación asociada a la mera exposición al cobro.

3. Estrategia Específica para el Segmento “Veteranos”

a) Diseño de membresía adaptada al segmento
Se propone desarrollar un plan específico para este segmento demográfico que incluya soporte técnico telefónico prioritario y atención personalizada.

El análisis evidenció una mayor tasa de abandono asociada a la ausencia de servicios de soporte. Incorporar este atributo como parte estructural de la propuesta de valor puede incrementar la percepción de acompañamiento, confianza y seguridad, factores especialmente relevantes en este grupo.

4. Estrategia de Fidelización Contractual

a) Garantía de estabilidad de precios
Se recomienda ofrecer a los clientes con modalidad “Mes a Mes” la opción de migrar a un contrato anual con garantía de mantenimiento del precio durante el período pactado.

La incertidumbre frente a eventuales incrementos tarifarios constituye un factor de riesgo en clientes sin compromiso contractual. La garantía de precio actúa como mecanismo de reducción de incertidumbre y estímulo a la permanencia.

5. Desarrollo de un Ecosistema de Protección y Valor Agregado

a) Bundles estratégicos de seguridad y soporte
Se sugiere incorporar, de manera promocional durante los primeros meses, servicios como “Seguridad Online” y “Soporte Técnico” en los planes básicos.

Una vez configurados e integrados en la rutina del usuario, estos servicios incrementan el costo de cambio percibido y fortalecen la dependencia funcional del cliente respecto a la compañía. Este enfoque contribuye a consolidar un ecosistema de servicios interrelacionados que dificulta la migración hacia competidores.
